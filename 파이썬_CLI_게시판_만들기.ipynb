```python
# Step 2. article list 명령어 수정중
# 구현해보니 최신글이 아래로 가는 상황
# 글이 많아지면 최신글 찾기가 더 번거로워지는 상황으로 예상됨
# 역으로 뒤집어서 순서를 최신글이 맨위에 오는 것으로
# 리스트를 뒤집으면 계속 뒤집히게 되어 출력을 뒤집는 방향으로
from datetime import datetime

print("== CLI 게시판 실행 ==")
i = 0 # 깔끔하게 하기 위해 수정
all_list = []
article_write = {} # 딕셔너리 추가

while True:
  cmd = input("명령어 ) ")



  if cmd == "exit":
    break


  elif cmd == "article write":
    i += 1; # 위치 이동 # di = 0
    a = input("제목 : ")
    b = input("내용 : ")
    print("{}번 글이 생성되었습니다".format(i))
    today = datetime.today()
    article_write = {"번호" : i, "제목" : a, "내용" : b, "날짜" : today.strftime('%Y-%m-%d %H:%M:%S')} # 날짜 추가
    all_list.append(article_write)



  elif cmd == "article list":
    print("====================")
    print("번호 / 제목 / 내용 / 날짜")
    # all_list.reverse() # 나오는 순서를 최신글부터 나오게 하기 위해서 # article list마다 뒤집히는 문제 발생
    for d in reversed(range(i)):
      print("{}   {}    {}    {}".format(all_list[d]["번호"], all_list[d]["제목"], all_list[d]["내용"], all_list[d]["날짜"])) # list 안 딕셔너리 값 호출하기
    print("===================")


  else:
    print("지원하지 않는 기능입니다")


print("== CLI 게시판 종료 ==")
```
