```python
# Step 2. article list 명령어 구현 중
# 딕셔너리 추가
# list 출력 형태 구현완료
# 출력시 마지막 기사만 출력됨
# 게시글 한개가 가지고 있는 여러 개의 값들을 헷갈리지 않고 어떤 순서로 저장했는지와 불러오기 위해 딕셔너리를 사용
# 게시글 여러 개(100개이상)는 리스트로 저장하는 것이 나아보임.

print("== CLI 게시판 실행 ==")
i = 1
name_list = []
content_list = []
article_write = {} # 딕셔너리 추가
all_list = []
while True:
  cmd = input("명령어 ) ")

  if cmd == "exit":
    break

  elif cmd == "article write":
    a = input("제목 : ") # 입력값을 받음
    b = input("내용 : ") # 입력값을 받음
    print("{}번 글이 생성되었습니다".format(i))
    article_write = {"번호" : i, "제목" : a, "내용" : b} # list 출력을 위한 딕셔너리 # 이것을 호출하면 최근 한개밖에 안불러지는 문제 발생
    all_list.append(article_write) # 게시글을 리스트로 한번에 저장
    i += 1;

  elif cmd == "article list":
    print("===============")
    print("번호 / 제목 / 내용")
    print("{}  / {}  / {}".format(article_write["번호"], article_write["제목"],article_write["내용"])) #기존 코드 # 지울 예정
    print(all_list) # 모든 게시글이 호출되는 것을 확인
    print("===============")


  else:
    print("지원하지 않는 기능입니다")


print("== CLI 게시판 종료 ==")
```
