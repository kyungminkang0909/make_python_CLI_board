```python
# Step 2. article list 명령어 수정중
# 불필요한 딕셔너리 선언 제거
# user-friendly - article list에서 article이 없는 경우 불필요한 표생성 대신 "게시글이 없습니다." 출력으로 변경

from datetime import datetime

print("== CLI 게시판 실행 ==")
i = 0
all_list = []
# article_write = {} # 딕셔너리 추가 # 지워도 잘 구동됨. # 삭제 예

while True:
  cmd = input("명령어 ) ")



  if cmd == "exit":
    break


  elif cmd == "article write":
    i += 1;
    a = input("제목 : ")
    b = input("내용 : ")
    print("{}번 글이 생성되었습니다".format(i))
    today = datetime.today()
    article_write = {"번호" : i, "제목" : a, "내용" : b, "날짜" : today.strftime('%Y-%m-%d %H:%M:%S')} # 날짜 추가
    all_list.append(article_write)



  elif cmd == "article list":
    if i == 0:
      print("게시글이 없습니다.")
    else:
      print("====================")
      print("번호 / 제목 / 내용 / 날짜")
      for d in reversed(range(i)):
        print("{}   {}    {}    {}".format(all_list[d]["번호"], all_list[d]["제목"], all_list[d]["내용"], all_list[d]["날짜"])) # list 안 딕셔너리 값 호출하기
      print("===================")


  else:
    print("지원하지 않는 기능입니다")


print("== CLI 게시판 종료 ==")
```
