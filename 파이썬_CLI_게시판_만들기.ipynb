```python
# Step 2. article list 명령어 수정중
# list 출력요소 날짜 추가
# article write 시 날짜 저장되게 함.


from datetime import datetime

print("== CLI 게시판 실행 ==")
i = 0 # 깔끔하게 하기 위해 수정
name_list = [] # 기존 코드 # 지울 예정
content_list = [] # 기존 코드 # 지울 예정
article_write = {} # 딕셔너리 추가
all_list = []
while True:
  cmd = input("명령어 ) ")

  if cmd == "exit":
    break

  elif cmd == "article write":
    i += 1; # 위치 이동 # di = 0
    a = input("제목 : ") # 입력값을 받음
    b = input("내용 : ") # 입력값을 받음
    print("{}번 글이 생성되었습니다".format(i))
    today = datetime.today()
    article_write = {"번호" : i, "제목" : a, "내용" : b, "날짜" : today.strftime('%Y-%m-%d %H:%M:%S')} # 날짜 추가
    all_list.append(article_write) # 게시글을 리스트로 한번에 저장


  elif cmd == "article list":
    print("====================")
    print("번호 / 제목 / 내용 / 날짜")
    for d in range(i):
      print("{}   {}    {}    {}".format(all_list[d]["번호"], all_list[d]["제목"], all_list[d]["내용"], all_list[d]["날짜c"])) # list 안 딕셔너리 값 호출하기
    print("===================")


  else:
    print("지원하지 않는 기능입니다")


print("== CLI 게시판 종료 ==")
```
